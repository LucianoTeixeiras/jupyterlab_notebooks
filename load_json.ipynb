# New Rows

## Import Libs

```python
import pandas as pd
import uuid as uid
from pandasql import sqldf
```

```python
pysqldf = lambda q: sqldf(q, globals())
```

## Read Files

```python
df = pd.read_json('raw_zone/pessoas.json')
```

```python
df.tail()
```

## Generated Uuid

```python
df['uuid'] = [str(uid.uuid4()) for _ in range(len(df.index))]
```

```python
df
```

```python
df = df.astype("object")
```

## Export To Parquet

```python
df.to_parquet('standar_zone/pessoas.parquet')
```

```python
df_delta = pd.read_parquet('standar_zone/pessoas.parquet')
```

```python
df_delta.to_parquet('standar_zone/pessoas_delta.parquet')
```

## PandasQL

```python
sql = """
    select 
         d.person_code
        ,d.person_name
        ,d.person_social_name
    FROM df d
    EXCEPT
    select 
         dd.person_code
        ,dd.person_name
        ,dd.person_social_name
    FROM df_delta dd
    """
```

```python
df_incremental = pysqldf(sql)
```

```python
df_incremental
```

```python
df_incremental['uuid'] = [str(uid.uuid4()) for _ in range(len(df_incremental.index))]
```

```python
df_incremental
```

```python
df_join = """
    select 
         d.person_code
        ,d.person_name
        ,d.person_social_name
        ,d.uuid
    FROM df d
    UNION ALL
    select 
         i.person_code
        ,i.person_name
        ,i.person_social_name
        ,uuid
    FROM df_incremental i
    """
```
